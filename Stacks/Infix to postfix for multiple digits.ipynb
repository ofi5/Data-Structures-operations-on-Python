```python
'''Taking an infix expression with multiple digits and converting it into postfix expression and evaluating the postfix expression.'''

#Creating class convert 
class convert():
    def __init__(self):
        self.stack=[]
        self.result=[]
        self.digits=[]
        self.evalu=[]
        self.precedence={'(':0, '-':1, '+':1, '*':2, '/':2, '^':3}
        
    def isEmpty(self):
        return self.stack==[]
    
    def push(self,ele):
        self.stack.append(ele)
            
    def pop(self):
        if not self.isEmpty():
            return self.stack.pop()
        else:
            return "&"
        
    def peek(self):
        if not self.isEmpty():
            return self.stack[-1]
        else:
            return "&"
    
    def infixToPostfix(self,exp):
        for c in exp:
            if c.isalnum():
                self.digits.append(c)
                
            elif c=='(':
                    self.stack.append(c)
                    
            elif c==')':
                self.result.append(''.join(self.digits))
                self.digits.clear()
                k=self.pop()
                while (k!='('):
                    self.result.append(k)
                    k=self.pop() 
                
            else:
                if self.digits!=[]:
                    self.result.append(''.join(self.digits))
                    self.digits.clear()
                while not self.isEmpty() and (self.precedence[self.peek()]>=self.precedence[c]):
                    self.result.append(self.pop())
                    
                self.push(c)
        
        if self.digits!=[]:
            self.result.append(''.join(self.digits))
            self.digits.clear()        
                
        while self.isEmpty()==False:
            self.result.append(self.pop())
            
        return self.result 
    
    def evaluate(self,exp):
        for i in exp:
            
            #print(i)
            if i.isalnum():
                self.evalu.append(i)
            else:
                a=self.evalu.pop()
                b=self.evalu.pop()
                self.evalu.append(str(eval(b+i+a)))
        return (self.evalu)        
                
                
        
               
c=convert()

u='(23+41+2*3)*3'

j=(c.infixToPostfix(u))

print(j)


print(''.join(j))

print(c.evaluate(j))
```
