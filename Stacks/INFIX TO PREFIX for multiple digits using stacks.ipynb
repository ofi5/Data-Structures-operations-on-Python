```python
'''to convert an infix expression to prefix and evaluate'''

# creating a class convert
class convert():
    def __init__(self):
        self.stack=[]
        self.result=[]
        self.digit=[]
        self.reverse=[]
        self.precedence={'(':0, '+':1, '-':1, '*':2,'/':2, '^':3}
        
    
    def isEmpty(self):
        return self.stack==[]
    
    def push(self,ele):
        self.stack.append(ele)
        
    def pop(self):
        if not self.isEmpty():
            return self.stack.pop()
      
    def peek(self):
        if not self.isEmpty():
            return self.stack[-1]
        
    def infixToPrefix(self,exp):
       #taking the infix expression and adding each elements in a new list self.result
        for c in exp:
            if c=='(':
                self.result.append(c)
            elif c.isdigit():
                self.digit.append(c)
            elif c==')':
                self.result.append(''.join(self.digit))
                self.digit.clear()
                self.result.append(c)
            else:
                if self.digit!=[]:
                    self.result.append(''.join(self.digit))
                    self.digit.clear()
                self.result.append(c)
                
        while self.digit!=[]:
            self.result.append(''.join(self.digit))
            self.digit.clear()
                
        print("the infix expression list is",self.result)
        
        # reversing the infix expression and adding to self.reverse list
        for l in self.result[-1::-1]:
            if l=='(':
                self.reverse.append(')')
            elif l==')':
                self.reverse.append('(')
            else:
                self.reverse.append(l)
        print("thereverseexpression list is:",self.reverse)
        
        #converting the self.reverse list to postfix expression
        self.result.clear()
        self.stack.clear()
        
        
        for g in self.reverse:
            if g.isalnum():
                self.result.append(g)
            elif g=='(':
                self.push(g)
            elif g==')':
                t=self.pop()
                while t!='(':
                    self.result.append(t)
                    t=self.pop()
            
            else:
                while not self.isEmpty() and self.precedence[self.peek()]>=self.precedence[g]:
                    self.result.append(self.pop())
                self.push(g)
                    
     
    
        while self.isEmpty()!=True:
            self.result.append(self.pop())
        print("postfix expression is",self.result)
        
        #postfix to prefix expression
        for h in self.result[-1::-1]:
            self.stack.append(h)
        print("the prefix expression is",self.stack)
        print(' '.join(self.stack))
        
        #evaluating the prefix expression
        self.digit.clear()
    
        for j in self.stack[-1::-1]:
            if j.isdigit():
                self.digit.append(j)
            else:
                a=self.digit.pop()
                b=self.digit.pop()
                self.digit.append(str(eval(a+j+b)))
        print("the valuation of prefix is:",self.digit)
                                      
                    
                
                
    
                
c=convert()
 
expr="(23+41-1*5)*8"

k=c.infixToPrefix(expr)
                


                        
                                      

```
