```python
'''To convert an Infix Expression to Prefix and evaluate the Prefix Expression'''

# creating a class convert
class convert():
    def __init__(self):
        '''
        Description: Constructor for class "convert".
        Input: None.
        Output: Creates 4 lists named stack, result, digit and reverse and 1 dictionary named precedence(setting precedence
                for parenthesis and operators).
        
        '''
        self.stack=[]
        self.result=[]
        self.digit=[]
        self.reverse=[]
        self.precedence={'(':0, '+':1, '-':1, '*':2,'/':2, '^':3}
        
    
    def isEmpty(self):
        ''' 
        Description: Function to check whether stack is empty.
        Input: Null.
        Output: Returns true if stack is empty else false.
        
        '''
        return self.stack==[]
    
    def push(self,ele):
        ''' 
        Description: Function to push an element on top of stack.
        Input: Element which is to be pushed.
        Output: Pushes the element on top of the stack.
        
        '''
        self.stack.append(ele)
        
    def pop(self):
        ''' 
        Description: Function to pop an element from stack.
        Input: Null.
        Output: Pops an element from top of the stack and returns the value.
        
        '''
        if not self.isEmpty():
            return self.stack.pop()
      
    def peek(self):
        ''' 
        Description: Function to peek an element from stack.
        Input: Null.
        Output: Peeks the top element of the stack and returns the value.
        
        '''
        if not self.isEmpty():
            return self.stack[-1]
        
    def infixToPrefix(self,exp):
        ''' 
        Description: Function to convert Infix Expression to Prefix Expression.
                     (Infix Expression is reversed and converted to postfix expression which is then reversed to get 
                     prefix expression.)    
        Input: Takes Infix expression.
        Output: Returns Prefix expression.
        
        '''
       #taking the infix expression and adding each elements in a new list self.result list
        for c in exp:
            if c=='(':
                self.result.append(c)
            elif c.isdigit():
                self.digit.append(c)
            elif c==')':
                self.result.append(''.join(self.digit))
                self.digit.clear()
                self.result.append(c)
            else:
                if self.digit!=[]:
                    self.result.append(''.join(self.digit))
                    self.digit.clear()
                self.result.append(c)
                
        while self.digit!=[]:
            self.result.append(''.join(self.digit))
            self.digit.clear()
                
        print("The Infix Expression list is",self.result)
        
        # reversing the infix expression and adding to self.reverse list
        for l in self.result[-1::-1]:
            if l=='(':
                self.reverse.append(')')
            elif l==')':
                self.reverse.append('(')
            else:
                self.reverse.append(l)
        print("The Reverse Infix Expression list is:",self.reverse)
        
        #converting the self.reverse list to postfix expression
        self.result.clear()
        self.stack.clear()
        
        
        for g in self.reverse:
            if g.isalnum():
                self.result.append(g)
            elif g=='(':
                self.push(g)
            elif g==')':
                t=self.pop()
                while t!='(':
                    self.result.append(t)
                    t=self.pop()
            
            else:
                while not self.isEmpty() and self.precedence[self.peek()]>=self.precedence[g]:
                    self.result.append(self.pop())
                self.push(g)
                    
     
    
        while self.isEmpty()!=True:
            self.result.append(self.pop())
        print("The Postfix Expression list is",self.result)
        
        #postfix to prefix expression
        for h in self.result[-1::-1]:
            self.stack.append(h)
        print("The Prefix Expression list is",self.stack)
        print("The FINAL Prefix Expression is:", ' '.join(self.stack))
        
        #evaluating the prefix expression
        self.digit.clear()
    
        for j in self.stack[-1::-1]:
            if j.isdigit():
                self.digit.append(j)
            else:
                a=self.digit.pop()
                b=self.digit.pop()
                self.digit.append(str(eval(a+j+b)))
        print("The Evaluation of Prefix is:",self.digit)
                                      
                    
                
                
    
                
c=convert() #Creating an object c of class "convert".
 
expr="(23+41-1*5)*8"

print("The Infix Expression is:",expr)

k=c.infixToPrefix(expr)
                


                        
                                      

```
