```python
'''To convert an Infix Expression to Postfix Expression using Stacks. (Single digit expressions)'''

#Creating a class "conversion".
class conversion():
    def __init__(self):
        '''
        Description: Constructor for class "conversion".
        Input: None.
        Output: Creates 2 lists named stack and result and 1 dictionary named precedence(setting precedence for parenthesis and operators).
        
        '''
        self.stack=[]
        self.result=[]
        self.precedence={'(':0, '-':1, '+':1,'*':2,'/':2, '^':3}
    
    
    def isEmpty(self):
        ''' 
        Description: Function to check whether stack is empty.
        Input: Null.
        Output: Returns true if stack is empty else false.
        
        '''
        return self.stack==[]
        
    
    def pop(self):
        ''' 
        Description: Function to pop an element from stack.
        Input: Null.
        Output: Pops an element from top of the stack and returns the value.
        
        '''
        if not self.isEmpty():
            return self.stack.pop()
     
    
   
    def peek(self):
        ''' 
        Description: Function to peek an element from stack.
        Input: Null.
        Output: Peeks the top element of the stack and returns the value.
        
        '''
        if not self.isEmpty():
            return self.stack[-1]

    
    def infixToPostfix(self,exp):
        
        ''' 
        Descrption: Function to convert infix expression to postfix expression. Loops each element of expression and checks for
                    alphanumeric charecter, checks for '(',')', checks precedence order in case of operators and returns 
                    postfix expression.
        Input: Takes Infix expression.
        Output: Returns Postfix expression.
        
        '''
        
        for i in exp:
            if i.isalnum():#alphanumeric charecter will be added to result list
                self.result.append(i)
                
            elif i=='(': #'(' will be added to stack if encountered while looping
                self.stack.append(i)
                 
            elif i==')': #when ')' is encounted stack elements are popped and added to result list until '(' occurs 
                #after which element is only popped but not added to stack
                k=self.pop()
                while(k!='('):
                    self.result.append(k)
                    k=self.stack.pop()
            else:#operator precedence is checked, if condition satisfies,element is popped from stack and added to stack
                #else element is added to stack
                while ((self.precedence[self.peek()]>=self.precedence[i]) and not self.isEmpty()):
                    j=self.stack.pop()
                    self.result.append(j)
                self.stack.append(i)
                
        #if stack has any more elements , they are popped and added to result stack        
        while self.isEmpty()==False:
            h=self.pop()
            self.result.append(h)
            
        return self.result# result list is returned     
            
c=conversion() #object c of class conversion is created 

exp="(2*3+4*(5-6))"
print("The infix expression is:",exp)

u= c.infixToPostfix(exp) #returns infix to postfix conversion in u

print("The postfix expression is:",''.join(u))#wil print the postfix result

        
```
