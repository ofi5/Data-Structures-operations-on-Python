```python
'''Operation on Stacks such as Push, Pop and Peek.'''


class stack():
    
    def __init__(self):
        '''
        Description: Constructor for class "stack".
        Input: None.
        Output: Creates a list named stack.
        
        '''
        self.stack=[]
        
    def emptylist(self):
        
        ''' 
        Description: Function to check whether stack is empty.
        Input: Null.
        Output: Returns true if stack is empty else false.
        
        '''
        return self.stack==[]
    
    def push(self,ele):
        ''' 
        Description: Function to push an element on top of stack.
        Input: Element which is to be pushed.
        Output: Pushes the element on top of the stack.
        
        '''
        self.stack.append(ele)
        
    def pop(self):
        ''' 
        Description: Function to pop an element from stack.
        Input: Null.
        Output: Pops an element from top of the stack and returns the value.
        
        '''
        if not stack.emptylist(self):
            return self.stack.pop()
        else:
            return(-1)
    def peek(self):
        ''' 
        Description: Function to peek an element from stack.
        Input: Null.
        Output: Peeks the top element of the stack and returns the value.
        
        '''
        if not stack.emptylist(self):
            return self.stack[-1]
        else:
            return(-1)
        

S=stack()        
while True:
    print("push")
    print("pop")
    print("peek")
    
    print("what do you want perform")
    
    do=input()
    
    if do=="push":
        print("enter element")
        ele=input()
        S.push(ele)
        print('****')
    elif do=="pop":
            ele=S.pop()
            if ele==-1:
                print("the list is empty")
                print("****")
            else:
                print(ele)
                print("****")
    elif do=="peek":
            ele=S.peek()
            if ele==-1:
                print("the list is empty")
                print("****")

            else:
                print(ele)
                print("****")
                
    else:
        print('wrong input')
        print("****")
        
                
                
        
```
