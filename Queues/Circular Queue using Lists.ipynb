```python
class node():
    def __init__(self,info,nxt=None):
        self.info=info
        self.nxt=nxt
        
class cirQueue():
    def __init__(self):
        self.front=self.rear=None
    
    def Enqueue(self,ele):
        cNode=node(ele)
        if self.front==None:
            self.front=cNode
            self.rear=cNode
            self.rear.nxt=cNode
        else:
            self.rear.nxt=cNode
            self.rear=cNode
            cNode.nxt=self.front
            
    def Dequeue(self):
        if self.front==None:
            print("Circular queue is Empty")
        elif self.front==self.rear:
            temp=self.front
            self.front=None
            self.rear=None
         
            print("the dequeued element is: ",temp.info)
            temp=None
        else:
            temp=self.front
            self.rear.nxt=temp.nxt
            self.front=temp.nxt
            print("the dequeued element is: ",temp.info)
            temp=None
            
    def display(self):
        if self.front==None:
            print("Circular queue empty is")
        else:
            current=self.front
            while current.nxt!=self.front:
                print(current.info)
                current=current.nxt
            print(current.info)
        
        
CQ=cirQueue()
CQ.Enqueue(1)
CQ.Enqueue(2)
CQ.Enqueue(3)
CQ.display()
print("----")
CQ.Dequeue()
CQ.display()
print('@@@@')
CQ.Dequeue()
CQ.display()
print("###")
CQ.Dequeue()
CQ.display()
print("%%%")
CQ.Dequeue()
CQ.display()
print("*******")
CQ.Dequeue()

    
```
