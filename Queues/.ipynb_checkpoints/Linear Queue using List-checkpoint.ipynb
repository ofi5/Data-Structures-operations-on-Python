```python
class node():
    def __init__(self,info,nxt=None):
        self.info=info
        self.nxt=nxt
        
class Queue():
    def __init__(self):
        self.front=self.rear=None
        
    def Enqueue(self,ele):
        newNode=node(ele)
        if self.front==None:
            self.front=newNode
            self.rear=newNode
        else:
            self.rear.nxt=newNode
            self.rear=newNode
            
    def Dequeue(self):
        if self.front==None:
            print("empty queue")
        elif self.front.nxt==None:
            temp=self.front
            self.front=None
            self.rear=None
            temp=None
        else:
            temp=self.front
            self.front=temp.nxt
            temp=None
            
    def display(self):
        current=self.front
        while current!=None:
            print(current.info)
            current=current.nxt
         
    
            
Q=Queue()

Q.Enqueue(10)
Q.Enqueue(20)
Q.Enqueue(30)
Q.display()
print("_______")
print("***")

Q.Dequeue()
Q.display()
print("***")
print("_______")

Q.Dequeue()
Q.display()
print("**")

Q.Dequeue()
Q.display()
print("*")

Q.Dequeue()

Q.display()

            
            
        
            
            
            
            
            
            
```
