```python
class node:
    def __init__(self,data):
        self.data=data
        self.right=None
        self.left=None
        
        
class BST:
    def newBst(self,root,ele):
        if root==None:
            return node(ele)
        if root.data>ele:
            root.left=self.newBst(root.left,ele)
        else:
            root.right=self.newBst(root.right,ele)
        return root    
            
    def inorder(self,root):
        stack=[]
        current=root
        while True:
            if current!=None:
                stack.append(current)
                current=current.left
            elif stack!=[]:
                current=stack.pop()
                print(current.data)
                current=current.right
            else:
                break
                
    def insuccessor(self,root):
        root=root.right
        while root.left!=None:
            root=root.left
        return root.data
    
    def inpredecessor(self,root):
        root=root.left
        while root.right!=None:
            root=root.right
        return root.data    
                
    def delete(self,root,key):
        if root==None:
            return None
        if root.data > key:
            root.left=self.delete(root.left,key)
        elif root.data < key:
            root.right=self.delete(root.right,key)
        else:
            if root.left==None and root.right==None:
                root=None
            elif root.right!=None:
                root.data=self.insuccessor(root)
                root.right=self.delete(root.right,root.data)
            else:#root.left!=None:
                root.data=self.inpredecessor(root)
                root.left=self.delete(root.left,root.data)
        return root     
            
                
root=None
b=BST()
elements=[10,5,2,25,20,30]
for ele in elements:
    root=b.newBst(root,ele)
    
b.inorder(root)    
print("_____")
b.delete(root,5)
print("_____")
b.inorder(root)

    
    
            
            
```
