```python
"""Creating a Single linked list, adding a node, performing operations such as insertion, deletion, traversal and displaying."""

#Creating a node using class
class node():
    
    def __init__(self,info,link=None):
        '''
        Description: Constructor for class node.
        Input: Information to be added in Node.
        Output: Creates a node with info and initialises link pointer to None.
    
        '''
        self.info=info
        self.link=link
        
#creating a linklist class        
class linkedlist():
    def __init__(self):
        '''
        Description: Constructor for class linkedlist.
        Input: Null.
        Output: Initialise head pointer to None.
        
        '''
            self.head=None
            
    #insertion of element at start of a linked list        
    def atstart(self,info):
        '''
        Description: Adding a node at start of list.
        Input: Information of node.
        Output: Adds a node with information at start of Single linked list.
        
        '''
        
        NewNode=node(info)
        if self.head==None:
            self.head=NewNode
        else:
            NewNode.link=self.head
            self.head=NewNode
    #insertion of element at end of Linked list        
    def atend(self,info):
        '''
        Description: Adding a node at end of list.
        Input: Information of node.
        Output: Adds a node with information at end of Single linked list.
        
        '''
    
        NewNode=node(info)
        if self.head==None:
            self.head=NewNode
        else:
            current=self.head
            while current.link!=None:
                current=current.link
            current.link=NewNode
            
    #displaying a linked list        
    def disp(self):
        '''
        Description: Display the list. 
        Input: None.
        Output: Displays the nodes of the list in ordered manner. 
        
        '''
        current=self.head
        while current != None:
            print(current.info)
            current=current.link
            
    #adding an element at certain index in a linked list        
    def addmid(self,info,index):
         '''
        Description: Adding a node except first and last position of list.
        Input: Information of node and index at which node has to be added.
        Output: Adds a node with information at given index of Single linked list.
        
        '''
        if index==0:
            return linkedlist.atstart(self,info)
        
        current=self.head
        count=1
        if index!=0:
            
            while current.link!=None:
                if count==index:
                    NewNode=node(info)
                    NewNode.link=current.link
                    current.link=NewNode
                count=count+1
                current=current.link 
        
        if count<=index:
            return linkedlist.atend(self,info)
            
        
        
    #deleting an element from linked list        
    def delete(self,ele):
        '''
        Description: Deleting node having element.
        Input: Element to be deleted.
        Output: Node having element(first instance) is deleted from the list.
        
        '''
        if self.head==None:
            Print("empty list")
            return
            
        if self.head.info==ele:
            temp=self.head
            self.head=temp.link
            temp=None
            return
            
        
        current=self.head
        while current.link!=None:
            if current.link.info==ele:
                temp=current.link
                current.link=temp.link
                temp=None
            current=current.link
            return
        
# creating a linkedlist object called LL.        
LL=linkedlist()

LL.atstart(5)
LL.atstart(2)
LL.atend(7)
LL.atend(8)
LL.atend(9)

LL.disp()

print("****")

LL.delete(5)
LL.delete(2)
LL.addmid(1,2)

LL.disp()
```
