```python
'''Creating a Circular Doubly Linked List and performing operations such as traversing,insertion,deletion'''

#Creating a node with forward and backward pointers.

class Node():
    '''
    Description: Constructor for class node.
    Input: Information to be added in Node.
    Output: Creates a node with info and initialises prev and nxt pointer to None.
    
    '''
    def __init__(self, info, prev=None, nxt=None):
        self.info=info
        self.prev=prev
        self.nxt=nxt

#Creating a class Circular doubly linked list        
class CLL():
    def __init__(self):
        '''
        
        Description: Constructor for class CLL.
        Input: Null(Constructor)
        Output: Initialises head pointer to None.
        
        '''
        self.head=None
    
    #Adding a node at start of list
    def atstart(self,info):
        '''
        Description: Adding a node at start of list.
        Input: Information of node.
        Output: Adds a node with information at start of Circular linked list.
        
        '''
        newNode=Node(info)
        #adding newnode in an empty list
        if self.head==None:
            self.head=newNode
            newNode.nxt=self.head
        
        #adding newnode at start when there are elements in the list
        current=self.head
        while current.nxt!=self.head:
            current=current.nxt
        current.nxt=newNode
        newNode.nxt=self.head
        self.head.prev=newNode
        self.head=newNode
    
    #Adding a node at end of list
    def atend(self,info):
        '''
        Description: Adding a node at end of list.
        Input: Information of node.
        Output: Adds a node with information at end of Circular linked list.
        
        '''
        newnode=Node(info)
        #adding newnode in an empty list
        if self.head==None:
            self.head=newnode
            newnode.nxt=self.head
        
        #adding newnode at end when there are elements in the list
        current=self.head
        while current.nxt!=self.head:
            current=current.nxt
        current.nxt=newnode
        newnode.nxt=self.head
        newnode.prev=current
    
    #adding new node except first and last position.
    def midd(self,info,index):
        '''
        Description: Adding a node except first and last position of list.
        Input: Information of node and index at which node has to be added.
        Output: Adds a node with information at given index of Circular linked list.
        
        '''
        newnode=Node(info)
        
        #if list is empty, node is added to the start of the list 
        if index==0:
            return CLL.atstart(self,info)
        
        current=self.head
        count=1
        while current.nxt!=self.head:
            if count==index:
                newnode.prev=current
                newnode.nxt=current.nxt
                current.nxt.prev=newnode
                current.nxt=newnode
            count=count+1
            current=current.nxt
        
        #if the index is greater than list length, the newnode as added at the end of the list.
        if count<=index:
            return CLL.atend(self,info)
        
    #deleting node from the circular list.        
    def delete(self,ele):
        '''
        Description: Deleting node having element.
        Input: Element to be deleted.
        Output: Node having element(first instance) is deleted from the list.
        
        '''
        #if element to be deleted is at the start of the list.
        if self.head.info==ele:
            temp=self.head
            current=self.head
            while current.nxt!=self.head:
                current=current.nxt
            current.nxt=temp.nxt
            self.head=temp.nxt
            temp=None
            return
        
        #if element to be deleted is anywhere else but the start of the list.
        current=self.head
        while current.nxt!=self.head:
            if current.nxt.info==ele:
                temp=current.nxt
                temp.nxt.prev=temp.prev
                current.nxt=temp.nxt
                temp=None
                
                return
            current=current.nxt        
    
    #Display the elements of the list.
    def display(self):
        '''
        Description: Display the list 
        Input: None
        Output: Displays the nodes of the list in ordered manner. 
        
        '''
        current=self.head
        while current.nxt!=self.head:
            print(current.info)
            current=current.nxt
        print(current.info)   
        
C=CLL() #Creating an object of class CLL.

C.atstart(10)
C.atstart(5)
C.atstart(2)

C.display()
print("#######")

C.atstart(6)
C.display() 
print("#######")

C.atstart(3)
C.atend(10)
C.midd(55,2)
C.midd(13,6)
C.midd(14,8)
C.midd(20,20)
C.midd(1,0)

C.display()
print("#######")  

C.delete(55)
C.delete(2)



C.display()

print("######")

C.delete(10)
C.display()
```
