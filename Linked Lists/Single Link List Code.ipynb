```python
#creating a node using class
class node():
    def __init__(self,info,link=None):
        self.info=info
        self.link=link
        
#creating a linklist class        
class linkedlist():
    def __init__(self):
            self.head=None
    #insertion of element at start of a linked list        
    def atstart(self,info):
        
        NewNode=node(info)
        if self.head==None:
            self.head=NewNode
        else:
            NewNode.link=self.head
            self.head=NewNode
    #insertion of element at end of Linked list        
    def atend(self,info):
    
        NewNode=node(info)
        if self.head==None:
            self.head=NewNode
        else:
            current=self.head
            while current.link!=None:
                current=current.link
            current.link=NewNode
            
    #displaying a linked list        
    def disp(self):
        current=self.head
        while current != None:
            print(current.info)
            current=current.link
            
    #adding an element at certain index in a linked list        
    def addmid(self,info,index):
        if index==0:
            return linkedlist.atstart(self,info)
        
        current=self.head
        count=1
        if index!=0:
            
            while current.link!=None:
                if count==index:
                    NewNode=node(info)
                    NewNode.link=current.link
                    current.link=NewNode
                count=count+1
                current=current.link 
        
        if count<=index:
            return linkedlist.atend(self,info)
            
        
        
    #deleting an element from linked list        
    def delete(self,ele):
        if self.head==None:
            Print("empty list")
            return
            
        if self.head.info==ele:
            temp=self.head
            self.head=temp.link
            temp=None
            return
            
        
        current=self.head
        while current.link!=None:
            if current.link.info==ele:
                temp=current.link
                current.link=temp.link
                temp=None
            current=current.link
            return
        
# creating a linkedlist object called LL.        
LL=linkedlist()

LL.atstart(5)
LL.atstart(2)
LL.atend(7)
LL.atend(8)
LL.atend(9)

LL.disp()

print("****")

LL.delete(5)
LL.delete(2)
LL.addmid(1,2)

LL.disp()
```
