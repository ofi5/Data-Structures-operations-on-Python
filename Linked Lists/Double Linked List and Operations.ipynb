```python
"""Creating a Double linked list,adding a node, performing operations such as insertion, deletion, traversal and displaying"""

#Creating a node for a double linked list
class Node():
    def __init__(self,ele, prev=None, nxt=None):
        self.ele=ele
        self.prev=prev
        self.nxt=nxt

#creating a class object Doublelinkedlist        
class DoubleLinkedList():
    def __init__(self):
        self.head=None
       
    #adding an element at the start of list
    def atstart(self,ele):
        newNode=Node(ele)
        #element is added when list is empty
        if self.head==None:
            self.head=newNode
            return
        #element is added at the start of list when list already has elements   
        if self.head!=None:
            newNode.nxt=self.head
            self.head.prev=newNode
            self.head=newNode
            return
    #adding element at the end of list        
    def atend(self,ele):
        newNode=Node(ele)
        #element is added to the list when the list is empty
        if self.head==None:
            self.head=newNode
            
        #element is added at the end of the list
        current=self.head
        while current.nxt!=None: 
            current=current.nxt
            
        current.nxt=newNode
        newNode.prev=current
        
    #element is added at the index location selected by the user    
    def midd(self,ele,index):
        newNode=Node(ele)
        #element is added to an empty list
        if self.head==None:
            self.head=newNode
            
        #element is added to start of list if the index is 0    
        if index==0:
            newNode.nxt=self.head
            self.head.prev=newNode
            self.head=newNode
            
        #element is added at the index chosen by user    
        current=self.head
        count=1
        while current.nxt!=None:
            if count==index:
                newNode.nxt=current.nxt
                current.nxt.prev=newNode
                current.nxt=newNode
                newNode.prev=current
               
            count=count+1
            current=current.nxt
        #if index is beyond the length of list, the element will be added to the end of list.    
        if count<=index:
            return DoubleLinkedList.atend(self,ele)
            
    #element is deleted by the user         
    def delete(self,ele):
        #first element is deleted if it matches
        if self.head.ele==ele:
            temp=self.head
            self.head=temp.nxt
            temp.nxt.prev=None
            temp=None
            return
        #element is deleted when matches with element    
        current=self.head
        while current.nxt!=None:
            if current.nxt.ele==ele:
                temp=current.nxt
                current.nxt=temp.nxt
                temp.nxt.prev=current
                temp=None
                return
            current=current.nxt
            
        #if element is not present in list,then it will return this.    
        print("element {} not present in list".format(ele)) 
        
        
    #display of elements in list    
    def display(self):
        current=self.head
        while current.nxt!=None:
            print(current.ele)    
            current=current.nxt 
        print(current.ele)

#Creating an object dll of class type DOublelinkedlist
dll=DoubleLinkedList()

dll.atstart(10)
dll.atstart(5)
dll.atstart(2)

dll.atend(20)
dll.atend(30)
dll.midd(15,8)
dll.display()

dll.delete(5)
print("####")
dll.display()

```
