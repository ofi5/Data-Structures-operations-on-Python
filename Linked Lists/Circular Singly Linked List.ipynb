```python
'''Creating a Circular singly linked list and performing operations such as insertion,deletion and traversing'''


#Creating a Node with forward pointer.


class Node():
    
    def __init__(self,info,nxt=None):
        
        '''
        Description: Constructor for class node.
        Input: Information to be added in Node.
        Output: Creates a node with info and initialises nxt pointer to None.
        
        '''
        self.info=info
        self.nxt=nxt

#Creating a class Circular single linked list(CSLL)        
class CSLL():
    def __init__(self):
        '''
        Description: Constructor for class CSLL.
        Input: Null.
        Output: Initialise head pointer to None. 
        
        '''
        self.head=None #head pointer pointing to none.
    
    #Creating a function to add node at start of the list
    def atstart(self,info):
        '''
        Description: Adding a node at start of list.
        Input: Information of node.
        Output: Adds a node with information at start of Circular singly linked list.
        
        '''
        
        newnode=Node(info)
        if self.head==None:#check:-if the list is empty
            self.head=newnode
            self.head.nxt=self.head
         
        #Traversing the list to check for link of last element to head pointer    
        current=self.head
        while current.nxt!=self.head:
            current=current.nxt
        #Adding a node at start and linking last element to make list cirular    
        current.nxt=newnode
        newnode.nxt=self.head
        self.head=newnode
    
    #Function to add node at end of list 
    def atend(self,info):
        '''
        Description: Adding a node at end of list.
        Input: Information of node.
        Output: Adds a node with information at end of Circular singly linked list.
        
        '''
        newnode=Node(info)
        if self.head==None:
            self.head=newnode
            newnode.nxt=self.head
        
        #Traversing the list to check for link of last element to head pointer 
        current=self.head
        while current.nxt!=self.head:
            current=current.nxt
        #Adding a node at end and linking last element to make list cirular    
        current.nxt=newnode
        newnode.nxt=self.head
    
    #Adding an element at index chosen by the user
    def atmid(self,info,index):
        '''
        Description: Adding a node except first and last position of list.
        Input: Information of node and index at which node has to be added.
        Output: Adds a node with information at given index of Circular linked list.
        
        '''
        newnode=Node(info)
        if self.head==None:
            self.head=newnode
            self.head.nxt=self.head
         
        #if index is 0,then element is added at start of list using inheritance
        if index==0:
            return CSLL.atstart(self,info)
        
        count=1
        current=self.head
        while current.nxt!=self.head:
            if count==index:
                newnode.nxt=current.nxt
                current.nxt=newnode
            count=count+1    
            current=current.nxt
        
        #if index is beyond the list length,then element is added at end of list using inheritance
        if count<=index:
            return CSLL.atend(self,info)
    
    
    # Deleting a node in CSLL
    def delete(self,ele):
        '''
        Description: Deleting node having element.
        Input: Element to be deleted.
        Output: Node having element(first instance) is deleted from the list.
        
        '''
        #if the first element has to be deleted
        if self.head.info==ele:
            current=self.head
            while current.nxt!=self.head:
                current=current.nxt
            current.nxt=self.head.nxt
            #self.head.info=None
            self.head=self.head.nxt
            return
        
        #to delete any other node except the first node
        current=self.head
        while current.nxt!=self.head:
            if current.nxt.info==ele:
                temp=current.nxt
                current.nxt=temp.nxt
                temp=None
                return
            current=current.nxt   
            
            
    #displaying the elements of Circular list    
    def display(self):
        '''
        Description: Display the list. 
        Input: None.
        Output: Displays the nodes of the list in ordered manner. 
        
        '''
        current=self.head
        while current.nxt!=self.head:
            print(current.info)
            current=current.nxt
        print(current.info)  
        
        
csll=CSLL() # Creating an object csll of class CSLL.

csll.atstart(10)

csll.display()

print("#####")

csll.atstart(5)
csll.atend(12)

csll.display()

print("#####")

csll.atstart(2)
csll.atend(15)
csll.atmid(1,0)
csll.atmid(3,2)
csll.atmid(4,3)
csll.atmid(17,8)
csll.atmid(16,8)
csll.atmid(18,13)
csll.atmid(11,6)

csll.display()

print("#####")

csll.delete(1)
csll.delete(18)

csll.display()
    
    
```
